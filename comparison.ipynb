```python
import pandas as pd

# Replace 'file1.csv' and 'file2.csv' with the paths to your CSV files
file1_path = 'ada_18_dead.csv'
file2_path = 'drmax_18_dead.csv'

# Reading the CSV files into dataframes
df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path)

# Now df1 and df2 contain the data from your CSV files
```

```python

# Extracting the 'prod_key' column from each dataframe
prod_key_df1 = set(df1['prod_key'])
prod_key_df2 = set(df2['prod_key'])

# Finding prod_keys present only in one dataframe but not in the other
prod_key_only_in_df1 = prod_key_df1 - prod_key_df2
prod_key_only_in_df2 = prod_key_df2 - prod_key_df1

# Printing the results
print("prod_keys only in df1(ada):", prod_key_only_in_df1)
print("prod_keys only in df2(drmax):", prod_key_only_in_df2)
```

```python
# Merge the dataframes on 'prod_key' column
# merged_df = pd.merge(df1, df2, on='prod_key', how='inner', suffixes=('_ada', '_drmax'))
# merged_df = merged_df.round(2)

# If you want to include prod_keys that are only in one dataframe, you can use 'outer' instead of 'inner'
merged_df = pd.merge(df1, df2, on='prod_key', how='outer',suffixes=('_ada', '_drmax'))
merged_df = merged_df.round(2)

# Display the merged dataframe
merged_df.head()
```

```python
# # Create new columns to indicate differences
merged_df['qty_diff'] = (merged_df['qty_ada'] != merged_df['qty_drmax'])
merged_df['local_acg_diff'] = (merged_df['local_acg_ada'] != merged_df['local_acg_drmax'])
merged_df['local_disc_diff'] = (merged_df['local_disc_ada'] != merged_df['local_disc_drmax'])

# Display the updated dataframe
merged_df.head()


# Calculate the differences and store them in new columns
merged_df['qty_diff_val'] = (merged_df['qty_ada'].fillna(0) - merged_df['qty_drmax']).fillna(0).round(2)
merged_df['local_acg_diff_val'] = (merged_df['local_acg_ada'].fillna(0) - merged_df['local_acg_drmax']).fillna(0).round(2)
merged_df['local_disc_diff_val'] = (merged_df['local_disc_ada'].fillna(0) - merged_df['local_disc_drmax']).fillna(0).round(2)

# Display the updated dataframez
merged_df.head()

```

```python
column_sums = merged_df.sum().round(2)


# Display the sums
print(column_sums)
```
