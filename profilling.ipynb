```python
import pandas as pd
import os
import glob

# Define the folder containing the CSV files
folder_path = 'profilling'  # Change this to your folder path

# Get a list of all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, '*.csv'))

# Loop through each CSV file
for file_path in csv_files:
    # Extract file name without extension
    file_name = os.path.basename(file_path).split('.')[0]
    
    # Extract table_name and analysis_date from the file name
    table_name = '_'.join(file_name.split('_')[:-1])
    analysis_date = file_name.split('_')[-1]
    
    # Load the CSV file
    df = pd.read_csv(file_path)
    
    # Add new columns
    df['table_name'] = table_name
    df['analysis_date'] = analysis_date
    
    # Define the output file path
    output_file_path = os.path.join(folder_path, file_name + '_modified.csv')
    
    # Save the modified DataFrame back to a CSV file
    df.to_csv(output_file_path, index=False)
    
    print(f"Processed and saved: {output_file_path}")

print("All files processed successfully.")
```

```python
import pandas as pd
import os
import glob

# Define the folder containing the modified CSV files
folder_path = 'profilling'  # Change this to your folder path

# Get a list of all modified CSV files in the folder
modified_csv_files = glob.glob(os.path.join(folder_path, '*_modified.csv'))

# Initialize an empty list to hold DataFrames
df_list = []

# Loop through each modified CSV file and append the DataFrame to the list
for file_path in modified_csv_files:
    # Ensure analysis_date is read as a string
    df = pd.read_csv(file_path, dtype={'analysis_date': str})
    df_list.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(df_list, ignore_index=True)

# Ensure analysis_date column is of string type
combined_df['analysis_date'] = combined_df['analysis_date'].astype(str)

# Optionally, save the combined DataFrame to a new CSV file
combined_file_path = os.path.join(folder_path, 'combined_data.csv')
combined_df.drop(columns=['Unnamed: 0'], inplace=True)
combined_df.to_csv(combined_file_path, index=False)

print(f"Combined DataFrame saved to: {combined_file_path}")
```
